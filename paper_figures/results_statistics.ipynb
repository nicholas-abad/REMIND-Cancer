```python
pcawg_results = pd.read_csv("/Users/nicholasabad/Desktop/workspace/data/_DATA_FOR_PAPER_/FinalResults_long.csv", delimiter="\t")
pcawg_results.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
pcawg_results = pcawg_results[(pcawg_results["REF"].str.len() == 1) & (pcawg_results["ALT"].str.len() == 1)].reset_index(drop=True)
pcawg_results
```

## Combine subset_long with EOPC-DE

```python
eopc_de = pd.read_csv("/Users/nicholasabad/Desktop/workspace/data/_DATA_FOR_PAPER_/SupplementaryTables/2_eopc_de_remaining_mutations_28Feb2024.csv")
eopc_de = eopc_de[(eopc_de["REF"].str.len() == 1) & (eopc_de["ALT"].str.len() == 1)].reset_index(drop=True)
```

```python
# NOTE: All columns were taken from EOPC-DE dataframe except for cn_score.

necessary_columns = [
    'score',
    'pid',
    'cohort',
    '#CHROM',
    'POS',
    'GENE',
    'SEQUENCE_CONTEXT',
    'REF',
    'ALT',
    'strand',
    'FPKM',
    'FPKM_Z_score',
    'within_cgc_list',
    'open_chromatin',
    'allele_frequency',
    'ICGC_Estimated_Purity',
    'num_created_tfs_passing_tf_expression_threshold',
    'created_tfs_passing_tf_expression_threshold',
    'num_destroyed_tfs_passing_tf_expression_threshold',
    'destroyed_tfs_passing_tf_expression_threshold',
    'num_remaining_tfs',
    'remaining_tfs',
    'JASPAR2020_CORE_vertebrates_non_redundant(tf_name,binding_affinity,seq1,seq2,raw,zscore,log,tf_sequence_logo)',
    'paths_with_recurrence(format=path,pid,cohort,bp,ref,alt,gene,chr,raw_score,zscore,log_score,confidence,purity,af,(tfs_seperated_by_//))',
    'num_original_mutations',
    'num_final_mutations',
    'num_promoter_mutations',
    'num_recurrent_mutations',
    'cn_score'
]
```

```python
combined_pcawg_and_eopc = pd.concat([pcawg_results, eopc_de], ignore_index=True)
combined_pcawg_and_eopc.sort_values("score", ascending=False, inplace=True)

combined_pcawg_and_eopc = combined_pcawg_and_eopc[necessary_columns]
```

```python
combined_pcawg_and_eopc.to_csv(
    "/Users/nicholasabad/Desktop/workspace/data/_DATA_FOR_PAPER_/pcawg_with_eopc_de_all_mutations_18March2025.csv",
    sep=",", index=False
)
```

# Why did the TERT C228T mutations drop out?

```python
tert_c228t = combined_pcawg_and_eopc.iloc[3]

tert_c228t_dataframe = pd.DataFrame()
tert_c228t_dataframe.loc[0, "pid"] = tert_c228t["pid"]
tert_c228t_dataframe.loc[0, "cohort"] = tert_c228t["cohort"]
tert_c228t_dataframe.loc[0, "zscore"] = tert_c228t["FPKM_Z_score"]


for idx, entry in enumerate(tert_c228t["paths_with_recurrence(format=path,pid,cohort,bp,ref,alt,gene,chr,raw_score,zscore,log_score,confidence,purity,af,(tfs_seperated_by_//))"].split(";")):
    path,pid,cohort,bp,ref,alt,gene,chr,raw_score,zscore,log_score,confidence,purity,af = entry.split(",")[:14]
    tert_c228t_dataframe.loc[idx+1, "pid"] = pid
    tert_c228t_dataframe.loc[idx+1, "cohort"] = cohort
    tert_c228t_dataframe.loc[idx+1, "zscore"] = zscore
    
pids_of_tert_c228ts_passing_pipeline = list(data[(data["POS"] == 1295228) & (data["GENE"] == "TERT")]["pid"].unique())

for idx, row in tert_c228t_dataframe.iterrows():
    tert_c228t_dataframe.loc[idx, "passes_pipeline"] = row["pid"] in pids_of_tert_c228ts_passing_pipeline
```

```python
# Result: 28 passed all filter steps.
combined_pcawg_and_eopc[(combined_pcawg_and_eopc["GENE"] == "TERT") & (combined_pcawg_and_eopc["POS"] == 1295228)].shape
```
